```python
from src.models.linreg import LinReg

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```


## Simple Linear Regression

```python
"""Create our initial Dataset"""
np.random.seed(69)
x = np.linspace(0, 10, 100)
y = 2 * x + 1 + np.random.normal(0, 1, 100)
simple_data = pd.DataFrame()
simple_data = (simple_data.assign(outcome=y, independent=x))
```

### Plot Data


```python
"""Plot the data"""
plt.figure(figsize=(10, 6))
plt.scatter(simple_data['independent'], simple_data['outcome'], alpha=0.7)
plt.title("Scatter Plot of y = 2x + 1 with Normal Noise")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
"""Create our Linear Regression Model"""

model = LinReg(df=simple_data, 
               outcome="outcome", 
               independent=["independent"],
               intercept=True,
               standard_error_type='non-robust')

fitted = model.fitted_values()
simple_data = simple_data.assign(fitted=fitted)
```

```python

"""Plot the data with the fitted values"""
"""Plot the data"""
plt.figure(figsize=(10, 6))
plt.scatter(simple_data['independent'], simple_data['outcome'], alpha=0.7)
plt.plot(simple_data['independent'], simple_data['fitted'], color='red', label='Fitted Line')
plt.title("Linear Regression Fit to the Data")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
"""Note this renders dynamic HTML styled with CSS to allow for interactivity in a notebook environment while working"""
model.summary()
```

### Multiple Linear Regression

```python
np.random.seed(69)
x = np.linspace(0, 5, 50)
z = np.linspace(3, 8, 50)
y = 1 + 3 * x + 5*z + np.random.normal(0, 2, 50)
multiple_data = pd.DataFrame({'outcome': y,
                     'independent1': x,
                     'independent2': z})

lm_robust = LinReg(df=multiple_data, outcome="outcome", independent=["independent1", "independent2"], standard_error_type='robust')
```

```python
lm_robust.summary()
```
